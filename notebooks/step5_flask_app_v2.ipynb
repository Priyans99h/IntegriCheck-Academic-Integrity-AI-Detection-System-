## Cell 3: Plagiarism Engine Quick Test
**Kya karta hai:** Plagiarism engine ke output ka structure verify karta hai — sab expected fields hain?

### Expected Output Fields:
| Field | Type | Purpose |
|-------|------|---------|
| `final_score` | float | 0-100 plagiarism % |
| `verdict` | str | "HIGH PLAGIARISM" etc. |
| `risk_level` | str | HIGH/MEDIUM/LOW/NONE |
| `word_count` | int | Input text word count |
| `tfidf_matches` | list | TF-IDF matched sources |
| `sbert_matches` | list | Semantic matches |
| `char_highlights` | list | Character positions for PDF |
| `match_categories` | dict | not_cited, missing_quotation etc. |

### Test Texts:
1. Academic text → Should score HIGH
2. Original text → Should score LOW/NONE

````python
import numpy as np
import re
from sklearn.metrics.pairwise import cosine_similarity

# ---------------------------------------------------
# Helper Functions
# ---------------------------------------------------

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()


def get_risk_level(score):
    if score >= 75:
        return "HIGH"
    elif score >= 40:
        return "MEDIUM"
    elif score >= 15:
        return "LOW"
    return "NONE"


def get_verdict(score):
    if score >= 75:
        return "HIGH SIMILARITY"
    elif score >= 40:
        return "MODERATE SIMILARITY"
    elif score >= 15:
        return "LOW SIMILARITY"
    return "ORIGINAL"


def detect_match_category(score):
    if score >= 75:
        return "direct-copy"
    elif score >= 40:
        return "paraphrased"
    elif score >= 15:
        return "minor-overlap"
    return "original"


def build_char_highlights(text, matches):
    """
    Creates character-level highlights
    """
    highlights = []

    lowered = text.lower()

    for m in matches[:5]:
        snippet = m.get("text", "")[:120].lower()

        # small phrases
        words = snippet.split()

        for w in words:
            if len(w) < 6:
                continue

            idx = lowered.find(w)

            if idx != -1:
                highlights.append({
                    "start": idx,
                    "end": idx + len(w),
                    "text": text[idx:idx+len(w)],
                    "type": "match"
                })

    return highlights


# ---------------------------------------------------
# MAIN FUNCTION
# ---------------------------------------------------

def analyze_plagiarism(text):

    global tfidf_vectorizer
    global tfidf_matrix
    global corpus_df
    global corpus_embeddings
    global sbert_model

    try:

        text = clean_text(text)

        if len(text.split()) < 5:
            return {
                "final_score": 0,
                "verdict": "TOO SHORT",
                "risk_level": "NONE",
                "word_count": len(text.split()),
                "tfidf_matches": [],
                "sbert_matches": [],
                "char_highlights": [],
                "match_categories": [],
                "_mode": "short-text"
            }

        # ---------------------------------------------------
        # TF-IDF SEARCH
        # ---------------------------------------------------

        query_vec = tfidf_vectorizer.transform([text])

        tfidf_scores = cosine_similarity(
            query_vec,
            tfidf_matrix
        ).flatten()

        top_tfidf_idx = np.argsort(tfidf_scores)[::-1][:5]

        tfidf_matches = []

        for idx in top_tfidf_idx:

            score = float(tfidf_scores[idx])

            # Ignore weak matches
            if score < 0.15:
                continue

            row = corpus_df.iloc[idx]

            tfidf_matches.append({
                "source": str(row.get("source", "unknown")),
                "score": round(score * 100, 2),
                "text": str(row.get("text", ""))[:300]
            })

        # ---------------------------------------------------
        # SBERT SEARCH
        # ---------------------------------------------------

        query_embedding = sbert_model.encode([text])

        sbert_scores = cosine_similarity(
            query_embedding,
            corpus_embeddings
        ).flatten()

        top_sbert_idx = np.argsort(sbert_scores)[::-1][:5]

        sbert_matches = []

        for idx in top_sbert_idx:

            score = float(sbert_scores[idx])

            # stricter filtering
            if score < 0.35:
                continue

            row = corpus_df.iloc[idx]

            sbert_matches.append({
                "source": str(row.get("source", "unknown")),
                "score": round(score * 100, 2),
                "text": str(row.get("text", ""))[:300]
            })

        # ---------------------------------------------------
        # FINAL SCORE
        # ---------------------------------------------------

        tfidf_max = max([m["score"] for m in tfidf_matches], default=0)
        sbert_max = max([m["score"] for m in sbert_matches], default=0)

        # weighted average
        final_score = round(
            (tfidf_max * 0.55) +
            (sbert_max * 0.45),
            2
        )

        # Clamp
        final_score = min(final_score, 100)

        # ---------------------------------------------------
        # MATCH CATEGORIES
        # ---------------------------------------------------

        match_categories = []

        if tfidf_matches:
            match_categories.append("lexical-overlap")

        if sbert_matches:
            match_categories.append("semantic-overlap")

        category = detect_match_category(final_score)

        if category not in match_categories:
            match_categories.append(category)

        # ---------------------------------------------------
        # CHAR HIGHLIGHTS
        # ---------------------------------------------------

        combined_matches = tfidf_matches + sbert_matches

        char_highlights = build_char_highlights(
            text,
            combined_matches
        )

        # ---------------------------------------------------
        # FINAL RESPONSE
        # ---------------------------------------------------

        result = {
            "final_score": final_score,
            "verdict": get_verdict(final_score),
            "risk_level": get_risk_level(final_score),
            "word_count": len(text.split()),
            "tfidf_matches": tfidf_matches,
            "sbert_matches": sbert_matches,
            "char_highlights": char_highlights,
            "match_categories": match_categories,
            "_mode": "hybrid"
        }

        return result

    except Exception as e:

        return {
            "final_score": 0,
            "verdict": "ERROR",
            "risk_level": "NONE",
            "word_count": len(text.split()),
            "tfidf_matches": [],
            "sbert_matches": [],
            "char_highlights": [],
            "match_categories": [],
            "_mode": "error",
            "error": str(e)
        }
````

## Cell 7: Flask API Integration Test (Threaded)
**Kya karta hai:** Flask server ko background thread mein start karta hai aur real HTTP requests bhejke test karta hai.

### How It Works:
```
Background Thread          Main Thread
─────────────────          ───────────
Flask server start    →    Wait 3 seconds
app.run(port=5001)    →    requests.post('/api/analyze/text')
                      →    Check response
                      →    Print results
```
