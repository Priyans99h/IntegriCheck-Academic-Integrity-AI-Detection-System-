# IntegriCheck — Step 1: Environment Setup

**Kya karenge:** Libraries install, folder verify, imports test karna.

> Har cell `Shift+Enter` se run karo — upar se neeche.

```python
# CELL 1 — Libraries install karo
# Kyu: Ye saari libraries poore project mein use hongi
# Expected output: [OK] library_name for each

import subprocess, sys

libraries = [
    'datasets', 'transformers', 'sentence-transformers',
    'scikit-learn', 'flask', 'flask-cors', 'fpdf2',
    'requests', 'beautifulsoup4', 'arxiv', 'datasketch',
    'nltk', 'torch', 'numpy', 'pandas', 'matplotlib',
    'seaborn', 'tqdm', 'wikipedia-api', 'joblib',
    'python-docx', 'PyPDF2', 'accelerate', 'evaluate'
]

for lib in libraries:
    r = subprocess.run([sys.executable, '-m', 'pip', 'install', lib, '-q'],
                       capture_output=True, text=True)
    print(f'  [{"OK" if r.returncode==0 else "FAIL"}] {lib}')

print('\n=== Installation complete ===')
```

```python
# CELL 2 — NLTK data download
# Kyu: Tokenization ke liye NLTK ko extra data chahiye
# Expected output: Downloaded: resource_name for each

import nltk
for r in ['punkt', 'stopwords', 'wordnet', 'averaged_perceptron_tagger', 'punkt_tab']:
    nltk.download(r, quiet=True)
    print(f'  Downloaded: {r}')
print('NLTK ready!')
```

```python
# CELL 3 — Library import test
# Kyu: Verify karo sab libraries import ho rahi hain
# Expected output: OK + version number for each, 0 failed at end

import importlib
test_libs = [
    ('numpy','numpy'), ('pandas','pandas'), ('sklearn','scikit-learn'),
    ('nltk','nltk'), ('torch','torch'), ('transformers','transformers'),
    ('sentence_transformers','sentence-transformers'),
    ('datasets','datasets'), ('flask','flask'),
    ('datasketch','datasketch'), ('arxiv','arxiv'),
    ('bs4','beautifulsoup4'), ('requests','requests')
]

ok, fail = 0, 0
for mod, name in test_libs:
    try:
        m = importlib.import_module(mod)
        v = getattr(m, '__version__', 'installed')
        print(f'  OK  {name:<30} {v}')
        ok += 1
    except ImportError as e:
        print(f'  ERR {name:<30} {e}')
        fail += 1

print(f'\nResult: {ok} OK | {fail} failed')
if fail == 0:
    print('Step 1 COMPLETE — Move to Step 2!')
```
